```python
from graphviz import Digraph
from IPython.display import display, Image
dot = Digraph(comment='计算图', format='png')
dot.graph_attr['rankdir'] = 'LR'
dot.edge_attr.update(arrowhead='vee', arrowsize='0.5')
dot.node('x', 'x', shape='box')
dot.node('mult', 'torch.mul', shape='diamond')
dot.node('y', 'y', shape='box')
dot.node('z', 'z', shape='box')
dot.node('exp', 'torch.exp', shape='diamond')
dot.node('b', 'b')
dot.edges([['b', 'mult'],['x','mult'],['mult','y'],['y','exp'],['exp','z']])
Image(dot.render('pic/compute_graph.gv'))
```

```python
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline
plt.rcParams['figure.figsize'] = (5.0, 5.0)

x=np.linspace(-10,10,500)
y=(x>=0)*x
plt.plot(x,y,color='blue') 
plt.title('ReLU')
plt.savefig("pic/ReLU.png")
plt.show()
```

```python
x=np.linspace(-10,10,500)
y=(x>=0)*x+0.1*(x<0)*x
plt.plot(x,y,color='blue', label=r"$\alpha = 0.1$")
plt.plot(x,(x>=0)*x+0.01*(x<0)*x,color='red', label=r"$\alpha = 0.01$")
plt.plot(x,(x>=0)*x,color='black', label=r"ReLU$")
plt.title('LeakyReLU')
plt.savefig("pic/LeakyReLU.png")
plt.legend()
plt.show()
```

```python
import scipy.special as sc

x=np.linspace(-10,10,500)
y=sc.ndtr(x)*x
plt.plot(x,y,color='blue', label=r"GELU")
plt.plot(x,x/(1+np.exp(-1*x)),color='red', label=r"SiLU")
plt.plot(x,np.log(1+np.exp(x)),color='black', label=r"Softplus")
plt.title('GELU')
plt.legend()
plt.savefig("pic/GELU.png")
plt.show()
```

```python
x=np.linspace(-10,10,500)
plt.plot(x,1/(1+np.exp(-1*x)),color='blue', label=r"Sigmoid(x)")
plt.plot(x,(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)),color='red', label=r"tanh(x)")
plt.legend()
plt.title('Sigmoid and tanh')
plt.savefig("pic/sigmoid.png")
plt.show()
```

```python
x=np.linspace(-10,10,500)
plt.plot(x,(x>6)*6+(x<0)*0+((x<=6) & (x>=0))*x,color='blue', label=r"ReLU6(x)")
plt.plot(x,x-(np.exp(x)-np.exp(-x))/(np.exp(x)+np.exp(-x)),color='red', label=r"Tanhshrink(x)")
plt.legend()
plt.title('ReLU6 and Tanhshrink')
plt.savefig("pic/relu6.png")
plt.show()
```
